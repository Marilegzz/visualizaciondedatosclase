```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

r=1
mean=[15,5]
cov=[[1,r],[r,1]]
x,y=x,y=np.random.multivariate_normal(mean,cov,400).T
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)

plt.hist(x = x,
         bins=10)
plt.title('x')
plt.show()

plt.subplot(1,2,2)
plt.hist(x=y,
        bins=15
        orientation=horizontal)
plt.title('y')

plt.subplot(2,2,4)
plt.scatter(x=x,
           y=y)
plt.title('Scatter plot')
plt.show()
```

```python
df=pd.read_excel('datasets/Canada.xlsx',
                sheet_name='Canada by Citizenship',
                skiprows=range(20),
                skipfooter=2,
                 engine='openpyxl'
                )

df.head(2)
```

```python
df.drop(['Type','Coverage','AREA','REG','DEV'],
       axis=1,
       inplace=True)
```

```python
df.rename(columns={
    'OdName': 'Country',
    'AreaName':'Continent',
    'RegName':'Region'
},
         inplace=True
         )
df.columns
```

```python
df.set_index('Country',
            inplace=True)
```

```python
df.head(2)
```

```python
df.columns
```

```python
df.columns=list(map(str, df.columns))  #poner como string los nombres de las columnas
df.columns
```

```python
df.head(2)
```

```python
df['Total']=df.iloc[:, 4:].sum(axis=1)
df.head(2)
```

```python
df.iloc[:,3:5]
```

```python
years=list(map(str,range(1980,2014)))
years
```

```python
df_continentes=df.groupby('Continent',
                         axis= 0,
                         ).sum()
```

```python
df_continentes
```

```python
df_continentes.sort_values(['Total'],
                          ascending=False,
                          axis=0,
                          inplace=True)
df_continentes.head()
```

```python
df_continentes['Total'].plot(kind='pie',
                           figsize=(5,6),
                           autopct= '%10.1f%%',
                           startangle=90,
                           shadow=True)
plt.title('Inmingration to Canada by Continent 1980-2013')
plt.axis('equal')#parece un circulo
plt.show()
```
